Import required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from random import shuffle
```

Input rows of article wordcount data; each row represents a single article position (1st, 2nd, etc) across all six columns

```python
%matplotlib inline
a = [151,428,176,473,274]
b = [59,958,309,379,205]
c = [78,0,161,612,151]
d = [102,0,60,0,220]
e = [53,0,222,0,249]
f = [54,0,270,0,223]
g = [100,0,0,0,0]
h = [70,0,0,0,0]
j = [80,0,0,0,0]
k = [34,0,0,0,0]
l = [108,0,0,0,0]
m = [72,0,0,0,0] 

```

Input rows of article wordcount data for second page

```python
n=[1741,1932,41,147,147]
o=[0,0,80,84,136]
p=[0,0,487,41,75]
q=[0,0,275,65,28]
r=[0,0,134,87,107]
s=[0,0,631,20,62]
t=[0,0,0,516,62]
u=[0,0,0,441,55]
v=[0,0,0,28,63]
w=[0,0,0,0,66]
x=[0,0,0,0,122]
y=[0,0,0,0,141]
z=[0,0,0,0,135]
```

Combine rows into a 'table', or list of lists. Each list represents a page of articles

```python
values_two = [a,b,c,d,e,f,g,h,j,k,l,m]
values = [n,o,p,q,r,s,t,u,v,x,y,z]
```

Establish maximum number of rows and columns in the data

```python
max_cols = len(values)
max_rows = len(values[0])
```

Establish the combined wordcount of articles per column

```python
values_sums = [sum([r[i] for r in values]) for i in range(max_rows)]
```

Normalise data (to create 100% columns) by dividing article word count by total column wordcount

```python
values_norm = [[v/values_sums[i] for i, v in enumerate(row)] for row in values]
```

Colourise the graph with random colours

```python
nn=[0,0,0,0,1]
oo=[0,0,0,0,2]
pp=[0,0,0,0,3]
qq=[0,0,0,0,4]
rr=[0,0,0,0,5]
ss=[0,0,0,0,6]
tt=[0,0,0,0,7]
uu=[0,0,0,0,8]
vv=[0,0,0,0,9]
ww=[0,0,0,0,10]
xx=[0,0,0,0,11]
yy=[0,0,0,0,12]
zz=[0,0,0,0,13]
values_colour = [nn,oo,pp,qq,rr,ss,tt,uu,vv,xx,yy,zz]
```

Create the column chart with "values"

```python
fig, ax = plt.subplots(1, figsize=(8, 16))
for row_num in range(max_cols):
    ax.bar(range(max_rows), values_norm[row_num], bottom=[sum([values_norm[i][j] for i in range(row_num)]) for j in range(max_rows)], width=1, edgecolor='#000000', color=[plt.get_cmap('Paired')(i) for i in values_colour[row_num] ], alpha=0.5)
ax.set_xlim(-0.5, max_rows-0.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylim(0, 1)
for x in range(max_rows):
    for y in range(max_cols):
        if values[y][x] != 0.0:
            ax.text(x, values_norm[y][x]/2 + sum([values_norm[newy][x] for newy in range(y)]), values[y][x], fontsize=10)
plt.savefig("value.pdf")
```

Re-run for second dataset

```python
max_cols = len(values_two)
max_rows = len(values_two[0])
values_two_sums = [sum([r[i] for r in values_two]) for i in range(max_rows)]
values_two_norm = [[v/values_two_sums[i] for i, v in enumerate(row)] for row in values_two]
colors = [plt.get_cmap('jet')(i) for i in np.linspace(0, 1, max_rows*max_cols)]
shuffle(colors)
fig, ax = plt.subplots(1, figsize=(8, 16))
for row_num in range(max_cols):
    ax.bar(range(max_rows), values_two_norm[row_num], bottom=[sum([values_two_norm[i][j] for i in range(row_num)]) for j in range(max_rows)], width=1, edgecolor='#000000', color=colors[row_num*max_rows:(row_num+1)*max_rows], alpha=0.5)
ax.set_xlim(-0.5, max_rows-0.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylim(0, 1)
for x in range(max_rows):
    for y in range(max_cols):
        if values_two[y][x] != 0.0:
            ax.text(x, values_two_norm[y][x]/2-.005 + sum([values_two_norm[newy][x] for newy in range(y)]), values_two[y][x], fontsize=10)
plt.savefig("value.pdf")
```
